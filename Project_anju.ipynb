1. **Course Registration System**

```python
'''Design a University Course Registration System.
The system should allow creation of courses with credits and seat limits, student
registration, and course enrollment.
Constraints include maximum credit limits per student, prevention of duplicate
registrations, and seat availability enforcement. Summary reports for students and courses
must be generated.'''
```

```python
courses = {}
students = {}

def add_course():
    cid = input("Enter Course ID: ")
    name = input("Enter Course Name: ")
    credits = int(input("Enter Credits: "))
    seats = int(input("Enter Seat Limit: "))

    if credits <= 0 or seats <= 0:
        print("Invalid credits or seats.")
        return

    courses[cid] = {
        "name": name,
        "credits": credits,
        "seats": seats,
        "enrolled": []
    }
    print("Course added successfully.")

def register_student():
    sid = input("Enter Student ID: ")
    name = input("Enter Student Name: ")
    max_credits = int(input("Enter Max Credit Limit: "))

    students[sid] = {
        "name": name,
        "max_credits": max_credits,
        "enrolled_courses": []
    }
    print("Student registered successfully.")

def enroll_student():
    sid = input("Enter Student ID: ")
    cid = input("Enter Course ID: ")

    if sid not in students or cid not in courses:
        print("Invalid student or course ID.")
        return

    student = students[sid]
    course = courses[cid]

    if cid in student["enrolled_courses"]:
        print("Student already enrolled in this course.")
        return

    if len(course["enrolled"]) >= course["seats"]:
        print("No seats available.")
        return

    current_credits = sum(courses[c]["credits"] for c in student["enrolled_courses"])
    if current_credits + course["credits"] > student["max_credits"]:
        print("Credit limit exceeded.")
        return

    student["enrolled_courses"].append(cid)
    course["enrolled"].append(sid)
    print("Enrollment successful.")

def student_report():
    sid = input("Enter Student ID: ")
    if sid not in students:
        print("Student not found.")
        return

    student = students[sid]
    print(f"\nStudent Name: {student['name']}")
    total_credits = 0

    for cid in student["enrolled_courses"]:
        course = courses[cid]
        print(f"- {course['name']} ({course['credits']} credits)")
        total_credits += course["credits"]

    print("Total Credits:", total_credits)

def course_report():
    cid = input("Enter Course ID: ")
    if cid not in courses:
        print("Course not found.")
        return

    course = courses[cid]
    print(f"\nCourse: {course['name']}")
    print("Credits:", course["credits"])
    print("Seats Available:", course["seats"] - len(course["enrolled"]))
    print("Enrolled Students:", course["enrolled"])

def main():
    while True:
        print("\n=== University Course Registration System ===")
        print("1. Add Course")
        print("2. Register Student")
        print("3. Enroll Student in Course")
        print("4. Student Summary Report")
        print("5. Course Summary Report")
        print("6. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            add_course()
        elif choice == "2":
            register_student()
        elif choice == "3":
            enroll_student()
        elif choice == "4":
            student_report()
        elif choice == "5":
            course_report()
        elif choice == "6":
            print("Exiting system.")
            break
        else:
            print("Invalid choice.")

main()
```

2. **Campus Store Inventory and Billing System**

```python
'''Develop a console-based Inventory and Billing System for a campus store.
The program should allow administrators to add products, update stock levels, and define
prices. Customers should be able to purchase multiple products in a single transaction.
The system must automatically generate bills showing item-wise cost, subtotal, tax, discount
(based on slabs), and final payable amount. Stock levels should update dynamically, and
low-stock alerts should be displayed.
Negative stock values must be prevented through validation logic. Use dictionaries for
inventory management and lists for cart handling.'''
```

```python
TAX_RATE = 0.05
LOW_STOCK_LIMIT = 5

inventory = {}

def add_or_update_product():
    pid = input("Enter Product ID: ")
    name = input("Enter Product Name: ")
    price = float(input("Enter Price: "))
    stock = int(input("Enter Stock Quantity: "))

    if stock < 0 or price < 0:
        print("Price or stock cannot be negative.")
        return

    inventory[pid] = {"name": name, "price": price, "stock": stock}
    print("Product added/updated successfully.")

def show_inventory():
    print("\n--- Inventory ---")
    for pid, data in inventory.items():
        print(f"{pid} | {data['name']} | ₹{data['price']} | Stock: {data['stock']}")
        if data["stock"] <= LOW_STOCK_LIMIT:
            print("Low Stock Alert!")

def calculate_discount(subtotal):
    if subtotal >= 1000:
        return subtotal * 0.15
    elif subtotal >= 500:
        return subtotal * 0.10
    return 0

def customer_purchase():
    cart = []

    while True:
        pid = input("Enter Product ID (or 'done' to finish): ")
        if pid.lower() == "done":
            break

        if pid not in inventory:
            print("Product not found.")
            continue

        qty = int(input("Enter Quantity: "))

        if qty <= 0:
            print("Quantity must be positive.")
            continue

        if qty > inventory[pid]["stock"]:
            print("Insufficient stock available.")
            continue

        cart.append((pid, qty))

    if not cart:
        print("Cart is empty.")
        return

    generate_bill(cart)

def generate_bill(cart):
    subtotal = 0
    print("\n--- BILL ---")
    print("Item\tQty\tPrice\tTotal")

    for pid, qty in cart:
        item = inventory[pid]
        cost = item["price"] * qty
        subtotal += cost
        inventory[pid]["stock"] -= qty
        print(f"{item['name']}\t{qty}\t₹{item['price']}\t₹{cost}")

    discount = calculate_discount(subtotal)
    tax = (subtotal - discount) * TAX_RATE
    total_payable = subtotal - discount + tax

    print("\nSubtotal:", subtotal)
    print("Discount:", discount)
    print("Tax:", tax)
    print("Final Amount Payable: ₹", round(total_payable, 2))

def admin_menu():
    while True:
        print("\n--- Admin Menu ---")
        print("1. Add/Update Product")
        print("2. View Inventory")
        print("3. Back to Main Menu")

        choice = input("Enter choice: ")

        if choice == "1":
            add_or_update_product()
        elif choice == "2":
            show_inventory()
        elif choice == "3":
            break
        else:
            print("Invalid choice.")

def main():
    while True:
        print("\n=== Campus Store System ===")
        print("1. Admin")
        print("2. Customer")
        print("3. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            admin_menu()
        elif choice == "2":
            customer_purchase()
        elif choice == "3":
            print("Thank you for using the system!")
            break
        else:
            print("Invalid choice.")

main()
```
